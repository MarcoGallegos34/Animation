```python


import plotly.graph_objects as go
import numpy as np

# Generate curve data
x = np.arange(0, 10, 0.1);
y = np.sin(x)
xm = np.min(x) 
xM = np.max(x) 
ym = np.min(y) 
yM = np.max(y) 
N = 50
xx = np.arange(0, 10, 0.1);
yy = np.sin(xx)


# Create figure
fig = go.Figure(
    data=[go.Scatter(x=x, y=y,
                     mode="lines",
                     line=dict(width=2, color="blue")),
          go.Scatter(x=x, y=y,
                     mode="lines",
                     line=dict(width=2, color="blue"))],
    layout=go.Layout(
        xaxis=dict(range=[xm, xM], autorange=False, zeroline=False),
        yaxis=dict(range=[ym, yM], autorange=False, zeroline=False),
        title_text="Simple Harmonic Motion", hovermode="closest",
        updatemenus=[dict(type="buttons",
                          buttons=[dict(label="Play",
                                        method="animate",
                                        args=[None])])]),
    frames=[go.Frame(
        data=[go.Scatter(
            x=[xx[k]],
            y=[yy[k]],
            mode="markers",
            marker=dict(color="red", size=10))])

        for k in range(N)]
)

fig.show()
```
